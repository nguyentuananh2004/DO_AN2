```python
import requests

# Định nghĩa các tham số cho yêu cầu API
latitude = 16.4637  # Vĩ độ của Huế, Việt Nam
longitude = 107.5909  # Kinh độ của Huế, Việt Nam
hourly_variables = [
    "temperature_2m",
    "relative_humidity_2m",
    "dew_point_2m",
    "precipitation_probability",
    "precipitation",
    "cloud_cover",
    "wind_speed_10m",
    "temperature_80m"
]

# Tạo URL yêu cầu
base_url = "https://api.open-meteo.com/v1/forecast"
params = {
    "latitude": latitude,
    "longitude": longitude,
    "hourly": ",".join(hourly_variables),
    "timezone": "auto"
}

# Gửi yêu cầu GET đến API
response = requests.get(base_url, params=params)

# Kiểm tra nếu yêu cầu thành công
if response.status_code == 200:
    data = response.json()
    # Xử lý dữ liệu ở đây
    print(data)
else:
    print(f"Yêu cầu thất bại với mã trạng thái: {response.status_code}")
```

```python
# Giả sử 'data' là dữ liệu JSON trả về từ API
hourly_data = data.get("hourly", {})
time = hourly_data.get("time", [])
temperature_2m = hourly_data.get("temperature_2m", [])
relative_humidity_2m = hourly_data.get("relative_humidity_2m", [])
dew_point_2m = hourly_data.get("dew_point_2m", [])
apparent_temperature = hourly_data.get("apparent_temperature", [])
precipitation_probability = hourly_data.get("precipitation_probability", [])
precipitation = hourly_data.get("precipitation", [])


# In ra nhiệt độ tại mỗi thời điểm
for t, temp, rh, dew, app_temp, precip_prob, precip in zip(time, temperature_2m, relative_humidity_2m, dew_point_2m, apparent_temperature, precipitation_probability, precipitation):
        print(f"Thời gian: {t}")
        print(f"Nhiệt độ (2 m): {temp}°C")
        print(f"Độ ẩm tương đối (2 m): {rh}%")
        print(f"Điểm sương (2 m): {dew}°C")
        print(f"Nhiệt độ biểu kiến: {app_temp}°C")
        print(f"Xác suất lượng mưa: {precip_prob}%")
        print(f"Lượng mưa: {precip} mm")
        print("-" * 40)
```

```python
import pandas as pd

# Giả sử 'data' là dữ liệu JSON trả về từ API
hourly_data = data.get("hourly", {})

# Tạo DataFrame từ dữ liệu
df = pd.DataFrame(hourly_data)

# Chuyển đổi cột 'time' thành định dạng datetime
df['time'] = pd.to_datetime(df['time'])
```

```python
# Xuất DataFrame thành tệp CSV
df.to_csv('du_lieu_thoi_tiet.csv', index=False, encoding='utf-8')
```
